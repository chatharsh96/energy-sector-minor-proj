```python
# detect outliers percentage in data, generally is greater than 5 % then needs treatment

# outliers = []
# for i in range(data.shape[1]):
#     min_t = data[data.columns[i]].mean() - (3 * data[data.columns[i]].std())
#     max_t = data[data.columns[i]].mean() + (3 * data[data.columns[i]].std())

#     count = 0
#     for j in data[data.columns[i]]:
#         if j < min_t or j > max_t:
#             count += 1
#     percentage = count / data.shape[0]
#     outliers[data.columns[i]] = "%.3f" % percentage

# outliers

outliers = {}
for column in data.columns:
    min_t = data[column].mean() - (3 * data[column].std())
    max_t = data[column].mean() + (3 * data[column].std())
    
    count = ((data[column] < min_t) | (data[column] > max_t)).sum()
    percentage = count / data.shape[0]
    outliers[column] = f"{percentage:.3f}"

print(outliers)
```

```python
# data rescaling, improves model accuracy, accelarate training process, normalize or standardize
# norm - values btw 0 and 1 , stand - all values mean 0 and stdev 1
x = data.iloc[:, 1:]
y = data.iloc[:, 0]

x = (x - x.min()) / (x.max() - x.min()) # normalization
x.head()
```

```python
# data splitting , train val test, larger dataset with million records or more 98:1:1,  smaller 60:20:20
# split by indexing or scikit learn


x.shape

```

```python
train_end = int(len(x) * 0.6)
val_end = int(len(x) * 0.8)
```

```python
x_shuffle = x.sample(frac=1, random_state=0)
y_shuffle = y.sample(frac=1, random_state=0)
```

```python
x_train = x_shuffle.iloc[:train_end,:]
y_train = y_shuffle.iloc[:train_end]
x_val = x_shuffle.iloc[train_end:val_end, :]
y_val = y_shuffle.iloc[train_end:val_end]
x_test = x_shuffle.iloc[val_end:, :]
y_test = y_shuffle.iloc[val_end:]
```

```python
model = nn.Sequential(nn.Linear(x_train.shape[1], 100), nn.ReLU(), nn.Linear(100,50), nn.ReLU(), nn.Linear(50, 25), nn.ReLU(), nn.Linear(25,1)).cuda()
```

```python
model
```

```python
loss_function = torch.nn.MSELoss()
```

```python
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
for i in range(10000):
    y_pred = model(x_train).squeeze()
    loss = loss_function(y_pred, y_train)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if i%100 == 0:
        print(i, loss.item())
```
